Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

Общий план работы: 
1. Предобработка данных;

3. Проведение исследовательского анализа данных;

•	airports_nearest — расстояние до ближайшего аэропорта в метрах (м)\

#### Заполнение нулями

Начнем со столбцов 'ponds_nearest', 'ponds_around_3000', 'parks_nearest', 'parks_around_3000'. Эти данные были собраны яндеком на основе данных, которые ввел пользователь. Но, видимо, здесь неправильно обработаны адреса или из-за каких-то технических проблем здесь отсутствуют данные. Здесь имеет смысл заменить данные на нули, т.к. эти данные в исследовании нам не понадобятся и суть не исказят.\
Пропуски в 'balcony' скорее всего показываетют, что балкона у квартиры нет, поэтому показатель можно заменить на ноль.

```python
fill_na = ['ponds_nearest', 'ponds_around_3000', 'parks_nearest', 'parks_around_3000', 'balcony']
df.isna().sum() #видим количество пропусков
```

Столбец 'is_apartment' состоит только из двух значений, но в данных присутствует только True, но в них не хватает False. Чтобы исправить, заменим эти данные.

```python
df['is_apartment'] = df['is_apartment'].fillna(False)
df[fill_na] = df[fill_na].fillna(0) #заполняем пропуски
```

```python
df.isna().sum() #проверка
```

После проверки видим, что от пропусков избавились.

#### Заполнение средними значениями

Со столбцами living_area, kitchen_area разберемся иначе.\
В данном случае мы можем примерно предположить, какая площать жилых помещений, если вычтем из общей площади кухню и предположительный процент санузла. Тоже самое можно проделать и для кухни, вычитая из общей площади жилые помещения и санузел. Так же проверим, есть ли в датасете варианты, где сумма площади кухни и жилых помещений совпадает или больше общей площади, т.к. вероятно, это аномалии, ведь в общую площадь должны входить еще и санузел и иные помещения, и уж тем более их сумма не может превышать общую площадь.

```python
df.loc[((df['living_area'] + df['kitchen_area']) >= df['total_area'])].count() #подсчет артефактов
```

Как видим таких данных в датасете немного, их можем спокойно удалить.

```python
df.loc[((df['living_area'] + df['kitchen_area']) >= df['total_area'])] = None
df = df.dropna(subset=['total_area']) #удаление через пустые значения
```

Так же еще в датасете присутствуют жилые площади, которые больше или очень близки к общей площади. Их тоже удаляем.

```python
df.loc[df['living_area'] >= df['total_area'] * 0.98] #подсчет
```

```python
df.loc[df['living_area'] >= df['total_area'] * 0.98] = None #удаление
df = df.dropna(subset=['total_area'])
```

```python
df.isna().sum() #проверка
```

```python
for row in df: #находим площадь кухни
    df.loc[df['kitchen_area'].isna() == True, 'kitchen_area'] =  df['total_area'] - df['living_area'] - (df['total_area'] * 0.08)
```

```python
for row in df: #находим площать жилых помещений
    df.loc[df['living_area'].isna() == True, 'living_area'] =  df['total_area'] - df['kitchen_area'] - (df['total_area'] * 0.08)
```

```python
df.isna().sum()
```

Таким образом получилось обработать не все строки, потому что значения где отсутствуют жилые помещения и кухня совпадают. Здесь  воспользуемся другим методом. Найдем среднюю долю кухни от общей площади помещения и попробуем угадать площадь кухни. Для жилых помещений проделаем то же самое. Дополнительно узнаем является ли данное помещение студией, т.к. у студий не отображается размер кухни, а только жилое помещение и иные.

```python
non_studio = df[df['studio'] == False] #список без студий
living_standarts = non_studio['living_area'].mean() / non_studio['total_area'].mean()
living_standarts #средняя доля жилых помещений в общей площади
```

```python
kitchen_standarts = non_studio['kitchen_area'].mean() / non_studio['total_area'].mean()
kitchen_standarts #средняя доля кухни в общей площади
```

Теперь подставим среднее значение площади кухни и жилых помещений, зная общую площадь.

```python
for row in df: #средняя площадь кухни
    df.loc[(df['kitchen_area'].isna() == True) & (df['studio'] == False), 'kitchen_area'] =  round((df['total_area'] * 0.92) * kitchen_standarts)
```

```python
for row in df: #средняя площадь жилых помещений
    df.loc[(df['living_area'].isna() == True) & (df['studio'] == False), 'living_area'] =  round((df['total_area'] * 0.92) * living_standarts)
```

Так как площадь иных помещений в студии мы не знаем, можно предположить, что она составляет в районе 0.92 от общей площади. Попробуем на неизвестные значения подставить данный коэффициент.

```python
for row in df: #средняя площадь жилых помещений в студии
    df.loc[(df['living_area'].isna() == True) & (df['studio'] == True), 'living_area'] =  (df['total_area'] * 0.92)
```

Для студий уберем площадь кухни и заменим на нули.

```python
df.loc[df['studio'] == True, 'kitchen_area'] = 0
```

Значения успешно поменялись.

В датафрейме присутствуют ошибки где сумма площади кухни/жилых/иных помещений больше, чем общая площадь. Найдем такие строки. Так как в данных отсутствует информация про иные виды помещений, попробуем предположить, что процент площади жилых помещений и кухни будет 0.92. Согласно этому будем фильтровать.

```python
area_round = df[['living_area', 'kitchen_area', 'total_area', 'studio']].copy() #отдельный список с нужными столбцами
```

```python
area_round.loc[((area_round['living_area'] + area_round['kitchen_area']) > area_round['total_area'] * 0.92)]
```

```python
#поиск артефактов
area_round.loc[((area_round['living_area'] + area_round['kitchen_area']) > area_round['total_area'])]
```

```python
area_round.loc[((area_round['living_area'] + area_round['kitchen_area']) == area_round['total_area'])] #площадь кухни+жилых равна общей площади
```

Все артефакты пропали и значения подправились. Остаётся немного строк с не очень правдоподобными данными. Я считаю что удалять эти строки не имеет смысла, т.к. в них содержится другая ценная информация.

Попробуем разобраться со столбцом 'ceiling_height'. Здесь отсутствует 40% значений. Чтобы заполнить эти пропуски можно попробовать подставить среднее значение потолков. Но этого будет недостаточно, т.к. в выборке предоставлены разные города и поселения, в которых виды домов различаются. Будет разумно подставить среднее значение потолка по населенному пункту.

```python
list_locality_name = df['locality_name'].unique() #список всех поселений
```

```python
median_ceiling_height = [] #список средних значений по населеным пунктам
for lists in list_locality_name:
    median_ceiling_height.append(df[df['locality_name'] == lists]['ceiling_height'].mean())
```

## Расчёты и добавление результатов в таблицу

### Цена квадратного метра

Для расчета цены квадратного метра, разделим в каждой строке цену на квадратный метр, а полученный результат добавим в таблицу.

```python
def sqr_meter_price(row): #формула для расчета
    meter = row['total_area']
    price = row['last_price']
    return price / meter
```

```python
df['sqr_meter_price'] = df.apply(sqr_meter_price, axis = 1).astype('int') # добавляем в таблицу новый столбец
```

```python
df.head() #контрольная проверка
```

Успешно добавили столбец.

```python
# check

df['sqr_meter_price'] = df['last_price'] / df['total_area']    
```

### День, месяц и год публикации объявления

Для добавления в таблицу отдельных столбцов с датой, месяцем и годом, воспользуемся специальным методом для типа datetime. Т.к. раньше мы дату уже перевели в данный тип, то осталось только добавить столбец.

```python
df['month_listed'] = df['first_day_exposition'].dt.month_name() #добавляем стобец с месяцев
df['weekday_listed'] = df['first_day_exposition'].dt.day_name() #с днем недели
df['year_listed'] = df['first_day_exposition'].dt.year #годом
```

```python
df.head() # проверяем
```

### Этаж квартиры
Чтобы добавить категорию этажа, сделаем сначала функцию для определения категории, а потом добавим в датафрейм методом append()

```python
def apartment_floor(row): # функция определения категории
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'
    
```

```python
df['apartment_floor'] = df.apply(apartment_floor, axis = 1) #добавляем в датафрейм
```

```python
df.head() #проверка
```

### Соотношение жилой и общей площади
Здесь можно сделать таким же способом как и с ценой за квадратный метр. Для каждого столбца сделаем но функции и добавим в таблицу.

```python
def kitchen_to_total_area(row): #соотношение кухни к общей площадей
    kitchen = row['kitchen_area']
    total = row['total_area']
    return kitchen / total
```

```python
def living_to_total_area(row): #соотношение жилой к общей площадей
    living = row['living_area']
    total = row['total_area']
    return living / total
```

```python
df['kitchen_to_total_area'] = df.apply(kitchen_to_total_area, axis = 1) # добавляем столбцы в таблицу
df['living_to_total_area'] = df.apply(living_to_total_area, axis = 1)
```

```python
df.head() #проверяем
```

### Задание 6

Изучим предложения квартир в Санкт-Петербурге. Задача — выяснить, какая область входит в центр. Создайте столбец с расстоянием до центра в километрах: округлите до целых значений. После этого посчитайте среднюю цену для каждого километра. Нарисуем график, и по нему выделим центр. 

Для начала выделим квартиры в Санкт_петербурге, для этого сделаем новый датафрейм, в который отфильтруем объявления по городу, плюс добавим условие, что расстояние до центра должно быть не меньше 1, т.к. в начале мы пропущеные значения заменили на ноль и приведем эти значения к целочисленному формату для удобства чтения.

```python
spb_supply = good_data[(good_data['locality_name'] == 'Санкт-Петербург') & (good_data['city_centers_nearest'] > 1)].copy()
spb_supply['city_centers_nearest_km'] = (spb_supply['city_centers_nearest'] / 1000).astype('int')
```

Теперь выведем среднюю цену для каждого километра.

```python
spb_supply_mean_price = spb_supply.pivot_table(index='city_centers_nearest_km', values='sqr_meter_price', aggfunc='mean').reset_index()
spb_supply_mean_price.head() #таблица со средними значениями на каждый км
```

```python
spb_supply_mean_price.sort_values(by='city_centers_nearest_km').plot(y='sqr_meter_price', x='city_centers_nearest_km', figsize=(18,4)); # диаграмма цены к км
```

По графику видим тренд к увеличению стоимости квадратного метра при приближении к центру. В районе 8 км происходит резкий ценовой скачек, поэтому с данного километра можем считать зону центральной.
